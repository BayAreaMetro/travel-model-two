# This notebook goes through preparing geojson shape inputs for shst match

```python
import geopandas as gpd
import geojson
from geojson import Feature, FeatureCollection
import json
from shapely.geometry import Point, LineString
import pandas as pd
```

```python
gtfs_folder = "../data/external/gtfs/muni/"
data_interim_folder = "../data/interim/muni_shst_shape/"
```

```python
shp_file_dict = {
            'muni' : gtfs_folder + "shapes.txt"}
```

```python
def gtfs_tojson_test(input_file, name):
    shape_df = pd.read_csv(input_file[name])
    geometry = [Point(xy) for xy in zip(shape_df.shape_pt_lon, shape_df.shape_pt_lat)]
    shape_df = gpd.GeoDataFrame(shape_df, geometry = geometry)
    
    line_df = shape_df.groupby(['shape_id'])['geometry'].apply(lambda x:LineString(x.tolist())).reset_index()
    line_gdf = gpd.GeoDataFrame(line_df, geometry = 'geometry')
    
    line_gdf.to_file(data_interim_folder + "muni.transit.geojson", 
                      driver='GeoJSON')
    
    return line_gdf
```

```python
muni_gdf = gtfs_tojson_test(shp_file_dict, 'muni')
```
